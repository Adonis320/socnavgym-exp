```python
import json
import matplotlib.pyplot as plt
import numpy as np

MAX_EP = 100_000

with open("./data/dqn_length.json", "r") as f:
    data = json.load(f)

plt.figure(figsize=(10, 6))

for i, item in enumerate(data):
    x = np.array(item["x"])
    y = np.array(item["y"])

    mask = x <= MAX_EP
    plt.plot(x[mask], y[mask], alpha=0.8, label=f"run {i}")

plt.xlabel("Episode")
plt.ylabel("Value")
plt.title("Performance up to 100k episodes")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
